```python
import numpy as np
import pandas as pd
import glob
import os
```

```python
myrusd_df = pd.read_csv("comodity-price-prediction-penyisihan-arkavidia-9/Mata Uang/MYRUSD=X.csv")
usdidr_df = pd.read_csv("comodity-price-prediction-penyisihan-arkavidia-9/Mata Uang/USDIDR=X.csv")

myrusd_df, usdidr_df
```

```python
myrusd_df =  myrusd_df[["Date", "Close"]]
myrusd_df = myrusd_df.rename(columns={"Close": "MYR/USD"})
myrusd_df
```

```python
usdidr_df = usdidr_df[["Date", "Close"]]
usdidr_df = usdidr_df.rename(columns={"Close": "USD/IDR"})
usdidr_df
```

```python
myridr_df = pd.merge(myrusd_df, usdidr_df, on="Date", how="inner")
myridr_df["MYR/IDR"] = myridr_df["MYR/USD"] * myridr_df["USD/IDR"]
myridr_df
```

```python
myridr_df.to_csv("comodity-price-prediction-penyisihan-arkavidia-9/to_idr.csv", index=False)
```
